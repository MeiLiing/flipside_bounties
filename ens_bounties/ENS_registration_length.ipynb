```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ens_registration_length.csv')
df.columns = ['type','wallet','avg_reg','count_reg']
df.head()
```

```python
plt.figure(figsize=(12,4))
ax = plt.subplot()
sns.boxplot(data=df, x='avg_reg')
plt.show()
```

```python
df[df.avg_reg > 100]
```

```python
plt.figure(figsize=(12,4))
ax = plt.subplot()
sns.scatterplot(data=df.loc[(df.avg_reg < 10) & (df.type == 'additional_domain')], x='avg_reg', y='count_reg', alpha = 0.3)
plt.show()
```
